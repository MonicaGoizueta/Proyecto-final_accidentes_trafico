# Selenium

```python
import requests
import pandas as pd
from time import sleep
import numpy as np


from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options




import warnings
warnings.filterwarnings('ignore')
```

```python
opciones= Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
#para ocultarme como robot
opciones.add_experimental_option('useAutomationExtension', False)
opciones.add_argument('--start-maximized') #empezar maximizado
opciones.add_argument('user.data-dir=selenium') #guarda las cookies
opciones.add_argument('--incognito')#incognito window
```

```python
# iniciamos el driver
driver = webdriver.Chrome(ChromeDriverManager().install())
## accedemos a la pagina web

driver.get("https://www.wunderground.com/history")
driver.implicitly_wait(10)
# esperamos
driver.find_element_by_css_selector("#truste-consent-button").click()
# aceptamos las cookies

sleep(5)
# seleccionamos la casilla para poner la ciudad por la que queremos buscar
driver.find_element_by_css_selector("#historySearch").send_keys("Madrid, Madrid, Spain", Keys.TAB)
sleep(5)
# por lo que sea le tenemos que dar dos veces al botón de enviar
driver.find_element_by_xpath('//*[@id="dateSubmit"]').click()
sleep(2)
driver.find_element_by_xpath('//*[@id="dateSubmit"]').click()
sleep(5)

driver.find_element_by_xpath('//*[@id="inner-content"]/div[2]/div[1]/div[1]/div[1]/div/lib-link-selector/div/div/div/a[3]').click()
# seleccionamos que nos los valores medios del mes
resultado = driver.find_element_by_css_selector("#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted").text

```

```python
url_list_1 = []

for month in range(1, 13):
    url_list_1.append(f"https://www.wunderground.com/history/monthly/LEMD/date/2021-{month}")
```

```python
result_list_1 = []

for i in url_list_1: 
    driver.get(i)
    try:
        sleep(5)
        driver.find_element_by_css_selector('#truste-consent-button').click()
        sleep(2)
        resultado = driver.find_element_by_css_selector("#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted").text
        result_list_1.append(resultado)
    except: 
        sleep(5)
        resultado = driver.find_element_by_css_selector("#inner-content > div.region-content-main > div.row > div:nth-child(5) > div:nth-child(1) > div > lib-city-history-observation > div > div.observation-table.ng-star-inserted").text
        result_list_1.append(resultado)
driver.quit()
```

```python
# hacemos un dataframe gordote
df_clima = pd.DataFrame()

for i in range(len(df_selenium)):
    x = list(df_selenium.iloc[i][0].split("\n"))
    
    mi_ind = x[1:].index("Max Avg Min")
    print(mi_ind)
    
    df_solo =  pd.DataFrame(np.array(x[1:]).reshape(7, mi_ind)).T
    
    df_solo.columns = ["Time", "Temperature (° F)", "Dew Point (° F)", "Humidity (%)", "Wind Speed (mph)", "Pressure (Hg)", "Precipitation (in)"]
    df_solo["month"] = x[1]
    df_clima = pd.concat([df_clima, df_solo], axis = 0)
```

```python
df_clima[["Tmax", "Avg", "Tmin"]] = df_clima["Temperature (° F)"].str.split(" ", expand = True)
```

```python
def separar_columnas(lista, columna):
    df_clima[lista] = df_clima[columna].str.split(" ", expand = True)
    return df_clima[columna].str.split(" ", expand = True)
```

```python
eliminar = ['Temperature (° F)', 'Dew Point (° F)', 'Humidity (%)',
       'Wind Speed (mph)', 'Pressure (Hg)']
```

```python
df_clima["mes"] = df_clima['month'].map({'Jan':1,
                             'Feb':2,
                             'Mar':3,
                            'Apr':4,
                            'May':5,
                            'Jun':6,
                             'Jul':7,
                             'Aug':8,
                            'Sep':9,
                             'Oct':10,
                             'Nov':11,
                            'Dec':12},
                             na_action=None)
```

```python
year = pd.datetime.now().year
df_clima['fecha']=pd.to_datetime({'year': 2021,
                           'month': df_clima["mes"],
                           'day': df_clima['dia']
                           })
```

```python
df_clima['fecha'] = pd.to_datetime(df_clima['fecha'])
```

```python
df_clima['temperatura(°C)_max'] = pd.to_numeric(df_clima["temperatura(°C)_max"], errors='coerce').astype('float64')
df_clima['temperatura(°C)_avg'] = pd.to_numeric(df_clima["temperatura(°C)_avg"], errors='coerce').astype('float64')
df_clima['temperatura(°C)_min'] = pd.to_numeric(df_clima["temperatura(°C)_max"], errors='coerce').astype('float64')
df_clima['pundo_de_condensacion(°C)_max'] = pd.to_numeric(df_clima["pundo_de_condensacion(°C)_max"], errors='coerce').astype('float64')
df_clima['pundo_de_condensacion(°C)_avg'] = pd.to_numeric(df_clima["pundo_de_condensacion(°C)_avg"], errors='coerce').astype('float64')
df_clima['pundo_de_condensacion(°C)_min'] = pd.to_numeric(df_clima["pundo_de_condensacion(°C)_min"], errors='coerce').astype('float64')
df_clima['humedad(%)_max'] = pd.to_numeric(df_clima["humedad(%)_max"], errors='coerce').astype('float64')
df_clima['humedad(%)_avg'] = pd.to_numeric(df_clima["humedad(%)_avg"], errors='coerce').astype('float64')
df_clima['humedad(%)_min'] = pd.to_numeric(df_clima["humedad(%)_min"], errors='coerce').astype('float64')
df_clima['presion(Hg)_max'] = pd.to_numeric(df_clima["presion(Hg)_max"], errors='coerce').astype('float64')
df_clima['presion(Hg)_avg'] = pd.to_numeric(df_clima["presion(Hg)_avg"], errors='coerce').astype('float64')
df_clima['presion(Hg)_min'] = pd.to_numeric(df_clima["presion(Hg)_min"], errors='coerce').astype('float64')
df_clima['vel_viento_max'] = pd.to_numeric(df_clima["vel_viento_max"], errors='coerce').astype('float64')
df_clima['vel_viento_avg'] = pd.to_numeric(df_clima["vel_viento_avg"], errors='coerce').astype('float64')
df_clima['vel_viento_min'] = pd.to_numeric(df_clima["vel_viento_min"], errors='coerce').astype('float64')
```
